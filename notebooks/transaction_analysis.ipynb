```python
import pandas as pd

transactions = pd.read_csv("../data/raw/transactions_train.csv")
transactions.head()
```

```python
# Check missing values
transactions.isnull().sum()

# Remove duplicate rows
transactions = transactions.drop_duplicates()

# Convert date column to datetime
transactions["t_dat"] = pd.to_datetime(transactions["t_dat"])

transactions.info()
```

```python
transactions["price"].describe()
```

```python
transactions["sales_channel_id"].value_counts()
```

```python
transactions["year"] = transactions["t_dat"].dt.year
transactions["month"] = transactions["t_dat"].dt.month

transactions.groupby("month").size()
```

```python
customer_features = transactions.groupby("customer_id").agg(
    Purchase_Count=("article_id", "count"),
    Total_Spending=("price", "sum"),
    Average_Purchase_Price=("price", "mean"),
    Last_Purchase_Date=("t_dat", "max")
).reset_index()

customer_features.head()
```

```python
latest_date = transactions["t_dat"].max()

customer_features["Recent_Activity_Days"] = (
    latest_date - customer_features["Last_Purchase_Date"]
).dt.days

customer_features.head()
```

```python
customer_features.describe()
```

```python
customer_features.isnull().sum()
```

```python
customer_features.sample(10)
```
